```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var_x = np.array([1.1,1.3,1.5,2.0,2.2,2.9,3.0,3.2,3.2,3.7,3.9,4.0,4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5, 6.8, 7])
var_y = np.array([39343, 46205, 37731, 43535, 39821, 56642, 60150, 54445, 64445, 57189, 63218, 55794, 56957, 57081, 61111, 67938, 73504, 79123, 83088, 81363, 93940, 91738, 98217, 101302,113812, 109431, 105582, 116969, 12635, 122391, 121872])
```



*   X stores the Prameter
*   Y stores the Target




```python
X = var_x.reshape(-1,1)
Y = var_y
```

```python
len(var_y)
```

```python
plt.scatter(var_x,var_y)
```

```python
df = pd.DataFrame({"Experience":var_x, "Salary":var_y})
```

```python
df.info()
```

Splitting the data into test and train data sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3)
```

```python
print(len(X_train), len(X_test))
```

Choose the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

Train the model on the training set

```python
model.fit(X_train, Y_train)
```

Make Predictions

```python
Y_test
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

The scatter plot below shows the visualization of the predicted values by the model trained (Orange) and the actual values of the dat set (purple)

```python
plt.scatter(X_test,Y_test)
plt.scatter(X_test,Y_pred)
```

Performance Matrix - MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test, Y_pred)
```

Accuracy

```python
model.score(X_test, Y_test)
```

*   The cross_val_score function maximizes scores, so it flips the sign of error metrics. A lower error (like 5) becomes -5, making it "better" than a higher error (like 9, which becomes -9).
*   Negative MSE (e.g., -5): Means your model is performing well; the actual MSE is 5.

*   Positive MSE (e.g., 5): Means your model is performing poorly; the errors are significant.
